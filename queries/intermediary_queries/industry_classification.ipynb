```python
def fetch_data_for_country(country):
    output_path = "Z:\\dati_moody\\data_processed\\firmographics_processed\\industry_classification\\{country}.parquet"
    os.makedirs(os.path.dirname(output_path.format(country=country)), exist_ok=True)
    query = f"""
        COPY (
            SELECT
                nace.bvd_id_number,
                nace.nace_rev_2_core_code_4_digits_
            FROM
                '{INDUSTRY_CLASSIFICATION_PATH}' AS nace
            WHERE
                nace.bvd_id_number IS NOT NULL
            AND
                nace.nace_rev_2_core_code_4_digits_ IS NOT NULL
            AND
                nace.bvd_id_number LIKE '{country}%'
        ) TO 
          '{output_path.format(country=country)}' (FORMAT PARQUET)
            """

    duckdb.sql(query)
```

## download data

```python
for country in ['RU']:
    print(f"Processing {country}...")
    fetch_data_for_country(country)
    print_memory_usage(f"After processing {country}: ")
```
