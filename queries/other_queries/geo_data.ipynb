```python
import pandas as pd
import duckdb
import os
```

```python
PATH = r"..\data_raw\firmographics\\contact_info\*.parquet"


def get_ownership_data(country, path=None):
        
    query = f"""
        SELECT 
            bvd_id_number, postcode, city, region_in_country, country_iso_code,
        FROM 
            '{path}'
        WHERE
            country_iso_code = '{country}'
    """

    conn = duckdb.connect()
    df = conn.execute(query).fetchdf()
    return df


def convert_to_parquet(df, output_path, country):
    os.makedirs(output_path, exist_ok=True)
    print(f"Converting to Parquet... - {country}")
    df.to_parquet(f"{output_path}/{country}.parquet")


country = "IT"
output_path = r"..\data_processed\\geo_data"
```

```python
countries_iso = pd.read_csv("..\data_raw\\others\\country_iso_code.csv")['Code'].tolist()
```

```python
test_path = "..\data_raw\\firmographics\\contact_info\\part-00000-54d921cc-a1c4-4ba5-941f-d71e06e76b2d-c000.snappy.parquet"

country = "US"
df = get_ownership_data(country, test_path)
convert_to_parquet(df, output_path, country)
```

```python
error_countries = []

for country in countries_iso:
    try:
        print(f"Processing country: {country}")
        # do only if country is not in outout path
        if os.path.exists(f"{output_path}/{country}.parquet"):
            continue
        df = get_ownership_data(country, PATH)
        convert_to_parquet(df, output_path, country)
    except:
        print(f"Error with country: {country}")
        continue


# save errors in json
error_countries = pd.DataFrame(error_countries)
error_countries.to_json(f"{output_path}/error_countries.json", orient='records')
```
