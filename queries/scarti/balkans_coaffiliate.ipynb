```python
import pandas as pd
import os

input_path = "Z:\\dati_moody\\zipped_files\\guos_eu"
output_path = "Z:\\dati_moody\\zipped_files\\balkans_coaffiliate"
os.makedirs(output_path, exist_ok=True)

years = list(range(2007, 2024))
balkans = ['BA', 'KV', 'AL', 'RS', 'ME', 'MK']

# for country in countries:
for country in ['CN']:
    for year in years:
        
        filename = f"{country}_{year}.csv.gz"
        infile = os.path.join(input_path, filename)

        # Skip if input file does not exist
        if not os.path.isfile(infile):
            print(f"Skipping missing file: {infile}")
            continue
        
        # Output file
        outfile = os.path.join(output_path, f"{country}_{year}_balkans.csv.gz")

        # # Skip if output already exists
        # if os.path.isfile(outfile):
        #     print(f"Skipping (already processed): {outfile}")
        #     continue
        
        print(f"Processing: {infile}")
        
        # Load file
        df = pd.read_csv(infile, compression='gzip', dtype=str)
        
        # Extract country code from subsidiary
        df['subsidiary_country'] = df['subsidiary_bvd_id'].str[:2]
        df['is_balkan'] = df['subsidiary_country'].isin(balkans)

        # Identify GUOs with at least one Balkan subsidiary
        guos_with_balkans = (
            df.groupby('guo_25')['is_balkan']
              .any()
              .reset_index()
        )
        guos_with_balkans = guos_with_balkans[guos_with_balkans['is_balkan']]

        # Filter the original dataframe
        df_balkans = df[df['guo_25'].isin(guos_with_balkans['guo_25'])]

        # Save output only if df_balkans has data
        if len(df_balkans) == 0:
            print(f"No Balkan subsidiaries for {country} {year}, skipping save.")
            continue
        
        df_balkans.to_csv(outfile, index=False, compression='gzip')
        print(f"Saved: {outfile}")
```
