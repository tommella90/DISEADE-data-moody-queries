```python
def create_query(input_path, country):
    query = f"""
        SELECT 
            * 
        FROM 
            read_parquet('{input_path}')
        WHERE 
            bvd_id_number like '{country}%'
    """
    return query

```

```python
def create_query(input_path, country):
    query = f"""
        SELECT 
            * 
        FROM 
            read_parquet('{input_path}')
        WHERE 
            guo_25 like '{country}%'
    """
    return query
```

```python
for input_rel_path, folder_name in paths:
    INPUT_PATH = BASE_PATH + input_rel_path
    FOLDER_NAME = folder_name

    for country in eu_27:
        new_file = convert_parquet_to_csv_gz(
            create_query=create_query, 
            folder_name=FOLDER_NAME, 
            input_path=INPUT_PATH,
            country=country
        )

        split_gz_file(
            input_gz=new_file,
            max_size_mb=50
        )
```
